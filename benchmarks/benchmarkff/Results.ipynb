# Benchmark: Comparsion of OpenFF performace in OpenMM and PELE

This benchmark replicates the method followed in *ChemRxiv. Preprint. 2020. https://doi.org/10.26434/chemrxiv.12551867.v2* to benchmark OpenFF and performs the same analysis using PELE and OpenMM in orther to compare both performances. 

We compare the energy-minimized geometries and energies for the different forcefields (*OpenFF-1.2.0*, *OpenFF-1.3.0* and *OPLS* the latter only for PELE) with respect to QM reference data by computing the following metrics:


 - Relative energy differences (ddE)
 - Root-mean-square deviation of atomic positions (RMSD)
 - Torsion fingerprint deviation (TFD)

## Relative energy differences (ddE)

### - Load output files from the analysis

```python
from numpy import genfromtxt

ddE_openmm = genfromtxt('results/ddE_openmm.csv', delimiter=',')
ddE_openmm13 = genfromtxt('results/ddE_openmm13.csv', delimiter=',')
ddE_pele = genfromtxt('results/ddE_pele.csv', delimiter=',')
ddE_pele13 = genfromtxt('results/ddE_pele13.csv', delimiter=',')
ddE_opls = genfromtxt('results/ddE_opls.csv', delimiter=',')
```

### - Comparative plots

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,6))
ax1.hist(-ddE_openmm , bins = 15, density = True, histtype = 'step', color = 'red', range = (-20,20), label = 'OpenMM with OpenFF-1.2.0')
ax1.hist(-ddE_openmm13 , bins = 15, density = True, histtype = 'step', color = 'gray', range = (-20,20), label = 'OpenMM with OpenFF-1.3.0')

ax2.hist(-ddE_pele , bins = 15, density = True, histtype = 'step', color = 'green', range = (-20,20), label = 'PELE with OpenFF-1.2.0')
ax2.hist(-ddE_pele13 , bins = 15, density = True, histtype = 'step', color = 'blue', range = (-20,20), label = 'PELE with OpenFF-1.3.0')
ax2.hist(-ddE_opls , bins = 15, density = True, histtype = 'step', color = 'orange', range = (-20,20), label = 'PELE with OPLS')

for ax in (ax1,ax2): 
    ax.set_xlabel('ddE(kcal/mol)', fontsize = 12)
    ax.set_ylabel('Density', fontsize = 12)
    ax.legend(loc="upper right")
    ax.axvline(x=0, ymax= 0.95,  ls = '--', lw = 0.7, markeredgewidth= 0.9,  color = 'gray')
    ax.set_ylim((0,0.25))
```

## Scatter plots of relative conformer energies versus TFD scores

```python
from numpy import genfromtxt
ddE_pele13_st = genfromtxt('results/ddE_pele13.csv', delimiter=',')
TFD_pele13_st = genfromtxt('results/TFD1_pele13.csv', delimiter=',')
```

```python
from numpy import genfromtxt
ddE_openmm13_st = genfromtxt('results/ddE_openmm13.csv', delimiter=',')
TFD_openmm13_st = genfromtxt('results/TFD1_openmm13.csv', delimiter=',')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import Normalize 
from scipy.interpolate import interpn

def density_scatter( x , y, title, ax = None, sort = True, bins = 40, **kwargs )   :
    """
    Scatter plot colored by 2d histogram
    """
    if ax is None :
        fig , ax = plt.subplots()
    fig.set_size_inches(8, 5)
    data , x_e, y_e = np.histogram2d( x, y, bins = bins, density = True )
    z = interpn( ( 0.5*(x_e[1:] + x_e[:-1]) , 0.5*(y_e[1:]+y_e[:-1]) ) , data , np.vstack([x,y]).T , method = "splinef2d", bounds_error = False)

    #To be sure to plot all data
    z[np.where(np.isnan(z))] = 0.0

    # Sort the points by density, so that the densest points are plotted last
    if sort :
        idx = z.argsort()
        x, y, z = x[idx], y[idx], z[idx]

    ax.scatter( x, y, c=z, s = 2,  **kwargs )
    ax.set_xlabel('TFD', fontsize = 12)
    ax.set_ylabel('ddE(kcal/mol)', fontsize = 12)
    ax.set_title(title,fontsize = 14)
    ax.set_yscale('symlog')
    ax.set_xlim([None,0.8])
    

    norm = Normalize(vmin = np.min(z), vmax = np.max(z))
    cbar = fig.colorbar(cm.ScalarMappable(norm = norm), ax=ax)
    cbar.ax.set_title('Density', fontsize = 10)
```

```python
density_scatter(TFD_openmm_st,-ddE_openmm_st, 'OpenMM with OpenFF-1.2.0')
density_scatter(TFD_openmm13_st,-ddE_openmm13_st, 'OpenMM with OpenFF-1.3.0')
density_scatter(TFD_pele_st,-ddE_pele_st, 'PELE with OpenFF-1.2.0')
density_scatter(TFD_pele13_st,-ddE_pele13_st, 'PELE with OpenFF-1.3.0')
density_scatter(TFD_opls_st,-ddE_opls_st, 'PELE with OPLS')
```
